```python
import numpy as np
import re
```

```python
def readFile(fileName):
    file = open(fileName,'r',encoding="cp437")
    fileText = ""
    for line in file:
        fileText += line
    return fileText
```

```python
def preProcess(text):
    # Remove non-letter chars
    text = re.sub("[^a-zA-Z ]"," ", text)
    # Change characters to lower
    text = text.lower()
    return text
```

```python
def preProcess_filtered(text):
  # Run the original preprocessing first
  text = preProcess(text)

  # Step 12: Remove words of length < 3 and > 8
  # \b matches word boundaries. [a-z] matches lowercase letters.
  # {1,2} catches lengths 1-2, {9,} catches lengths 9 and above.
  text = re.sub(r'\b[a-z]{1,2}\b|\b[a-z]{9,}\b', ' ', text)

  # Clean up extra spaces left behind
  text = re.sub(r'\s+', ' ', text).strip()

  return text
```

```python
def genReferenceWordList(texts,stopWords):
    # concatenate the texts
    allText = ""
    for line in texts:
        allText += line
    # Generate a word list
    wordsList =  allText.split()
    # Generate a word set
    wordsSet =  set(wordsList)
    # Remove the stop words from the word list
    stopWordsList = stopWords.split()
    stopWordsSet = set(stopWordsList)
    refWordSet = wordsSet.difference(stopWordsSet)
    return list(refWordSet)
```

```python
def vecDist(vec1,vec2):
    vecDiff = vec1-vec2
    # compute the distance ("pitagoras")
    vecSqr =  np.square(vecDiff)
    vecSum =  np.sum(vecSqr)
    return np.sqrt(vecSum)
```

```python
def getWordFrequency(texts,dictList):
    dictSize = len(dictList)
    nTexts = len(texts)
    wordFreq = np.empty((nTexts,dictSize))
    for i in range(nTexts):
        print("text" + str(i))
        for j in range(dictSize):
            wordFreq[i,j] = texts[i].count(dictList[j])
    return wordFreq
```

```python
# 1. Manhattan Distance: D = sum(|x - y|)
def vecDist_manhattan(vec1, vec2):
    vecDiff = vec1 - vec2
    return np.sum(np.abs(vecDiff))

# 2. Square Root Distance: D = sqrt(sum(|x - y|))
def vecDist_sqrt(vec1, vec2):
    vecDiff = vec1 - vec2
    # The sum happens FIRST, then the square root is applied to the total
    return np.sqrt(np.sum(np.abs(vecDiff)))

# Compute the distance matrices for the new functions
# (Assuming 'rows' and 'wordFreq' are already defined from your previous cells)
dist_manhattan = np.zeros((rows, rows))
dist_sqrt = np.zeros((rows, rows))

for i in range(rows):
    for j in range(rows):
        dist_manhattan[i,j] = round(vecDist_manhattan(wordFreq[i], wordFreq[j]))
        dist_sqrt[i,j] = round(vecDist_sqrt(wordFreq[i], wordFreq[j]))

print("Manhattan dist matrix = \n")
print(dist_manhattan)

print("\nSqrt dist matrix = \n")
print(dist_sqrt)
```
